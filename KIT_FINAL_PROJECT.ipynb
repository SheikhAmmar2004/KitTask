```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Audio=pd.read_csv('/content/audible_uncleaned.csv')
```

```python
Audio
```

```python
Audio=Audio.drop_duplicates()
```

```python
Audio.info()
```

```python
Audio.describe()
```

```python
Audio.isnull().value_counts()
```

Cleaning Author column


```python
Audio['author'].value_counts()
```

```python
Audio['author']=Audio['author'].str.replace('Writtenby:','')
```

```python
Audio['author']=Audio['author'].str.replace('NaN','Unknown')
```

```python
Audio['author']=Audio['author'].str.replace('矢島雅弘,','Masahiro Yajima,')
Audio['author']=Audio['author'].str.replace('中西貴之,','Takayuki Nakanishi,')
Audio['author']=Audio['author'].str.replace(',石橋遊','Ishibashi Tour')
```

```python
Audio
```

Cleaning narrator column


```python
Audio['narrator'].value_counts()
```

```python
Audio['narrator']=Audio['narrator'].str.replace('Narratedby:','')
```

```python
Audio['narrator']=Audio['narrator'].str.replace('NaN','Unknown')
```

```python
Audio['narrator']=Audio['narrator'].str.replace('矢島雅弘','Masahiro Yajima')
Audio['narrator']=Audio['narrator'].str.replace(',石橋遊',', Stone Bridge Tour')
Audio['narrator']=Audio['narrator'].str.replace('中西貴之','Takayuki Nakanish')
```

```python
Audio
```

Cleaning release date column


```python
Audio['releasedate']=pd.to_datetime(Audio['releasedate'])
```

```python
Audio
```

Cleaning language column


```python
Audio['language']=Audio['language'].str.replace('NaN','Unknown')
```

```python
Audio['language'].value_counts()
```

Cleaning the star column


```python
Audio['stars']=Audio['stars'].str.replace('out of 5','')
```

```python
Audio[['stars','ratings']]=Audio['stars'].str.split('stars',1,expand=True)
```

```python
Audio['stars']=Audio['stars'].str.replace('Not rated yet','0')
```

```python
Audio['stars']=Audio['stars'].astype(float)
```

```python
Audio
```

Cleaning the ratings column


```python
Audio['ratings']=Audio['ratings'].str.replace('ratings','')
```

```python
Audio['ratings']=Audio['ratings'].str.replace('rating','')
```

```python
Audio['ratings']=Audio['ratings'].str.replace(',','')
```

```python
Audio['ratings']=Audio['ratings'].fillna(0)
```

```python
Audio['ratings']=Audio['ratings'].astype(float)
```

```python
Audio
```

```python
Audio.info()
```

Cleaning the price column


```python
Audio['price']=Audio['price'].replace('Free','0')
```

```python
Audio['price'] = Audio['price'].str.replace(',', '').astype(float)
```

```python
Audio['price']=Audio['price'].fillna(0)
```

```python
Audio
```

```python
Audio=Audio.rename(columns=({'price':'price in $'}))
```

```python
Audio
```

Cleaning the time column

```python
Audio['time'] = Audio['time'].str.replace('hrs and', 'H').str.replace('hrs','H').str.replace('hr and', 'H').str.replace('mins', 'M').str.replace('min', 'M')
```

```python
Audio['time'] = Audio['time'].replace('Lessthan', pd.NaT)
```

```python
Audio['time'].value_counts()
```

```python
Audio['time'] = Audio['time'].replace('Lessthan', pd.NaT)
```

```python
Audio['time'] = pd.to_timedelta(Audio['time'], errors='coerce')
```

```python
Audio['time'] = Audio['time'].astype(str).str[7:]
```

```python
Audio=Audio.rename(columns=({'time':'duration'}))
```

```python
Audio
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1, 2, 2)
sns.scatterplot(data=Audio, x='stars', y='ratings',color='red')
plt.title('Stars vs. Ratings')
plt.tight_layout()
plt.show()
```

```python
Author_Counts = Audio['author'].value_counts()
Author_Counts = Author_Counts[Author_Counts >= 200]

plt.figure(figsize=(10, 8))
Author_Counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Author Names', fontsize=12)
plt.ylabel('Number of Audio Books', fontsize=12)
plt.title('Top Authors with Most Published Audio Books', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=Audio, x='releasedate', bins=50, kde=True,color='#008080')
plt.title('Distribution of Audio Books Released Over Time')
plt.xlabel('Release Date')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
Narrator_Counts = Audio['narrator'].value_counts()
Narrator_Counts = Narrator_Counts[Narrator_Counts >= 300]

plt.figure(figsize=(10, 8))
Narrator_Counts.plot(kind='bar', color='#FA8072', edgecolor='black')
plt.xlabel('Narrator Names', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Top Narrators with Most Published Audio Books', fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
Language_Counts = Audio['language'].value_counts()
Language_Counts = Language_Counts[Language_Counts >= 2000]

plt.figure(figsize=(10, 8))
Language_Counts.plot(kind='barh', color='purple', edgecolor='black')
plt.xlabel('Count', fontsize=12)
plt.ylabel('Languages', fontsize=12)
plt.title('Top Lanugages', fontsize=14)
plt.yticks(fontsize=10)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(16,5))
plt.subplot(1, 2, 1)
sns.scatterplot(data=Audio, x='price in $', y='ratings',color='#DC143C')
plt.title('Price vs. Ratings')
```
